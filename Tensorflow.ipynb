```python
#tensor aggregation

import numpy as np
tensor=tf.constant(np.random.randint(0,100,size=50))
tensor
```

```python
tf.reduce_min(tensor),tf.reduce_max(tensor),tf.reduce_mean(tensor)
```

```python
tf.reduce_sum(tensor)
```

```python
import tensorflow_probability as tfp
print(tfp.stats.variance(tensor))
#tf.math.reduce_variance(tensor)
#it doesnt work since it requires real or complex number so....
tf.math.reduce_std(tf.cast(tensor,dtype=tf.float16))
```

```python
tf.argmax(tensor)
#intution behind the positional maximum and minimum
tf.argmin(tensor)
```

```python
tensor[tf.argmin(tensor)]==tf.reduce_min(tensor)
```
